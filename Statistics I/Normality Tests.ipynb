```python
# Importing libs
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Creating variable with function rvs (100 elements) from a normal distribution data
data = norm.rvs(size = 1000)
data
```

```python
# Histogram
plt.hist(data, bins = 20)
plt.title('Data')
```

```python
# Generating chart to verify if the distribution is normal
fig, ax = plt.subplots()
stats.probplot(data, fit=True, plot=ax)
plt.show()
```

```python
# Execution of Shapiro Test
stats.shapiro(data)
```

```python
# Not normal data
data2 = skewnorm.rvs(4, size=1000)
```

```python
# Histogram
plt.hist(data2, bins = 20)
plt.title('Data')
```

```python
# Generating chart to verify if the distribution is normal
fig, ax = plt.subplots()
stats.probplot(data2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(data2)
```
